```python
import pandas as pd
import geopandas
import matplotlib.pyplot as plt
import numpy as np

from Geographycal_functions import localize_tweets
from Geographycal_functions import drop_non_geolocalised
from shapely.geometry import Point
```

# Donald Trump

```python
# import the raw data
data_donald = pd.read_csv("hashtag_donaldtrump.csv", lineterminator='\n')

data_donald = drop_non_geolocalised(data_donald, "lat", "long")
geo_donald = localize_tweets(data_donald, "World Trump data distribution")
```

**drop all the tweets localised outside the USA**

```python
# upload the USA map
usa = geopandas.read_file("Maps/states.shp")
print(usa)
```

```python
# get the border
#usa_borders = usa.boundary
# join our points in the geopandas dataframe and assign the state
geo_donald_mod = geopandas.sjoin(usa, geo_donald, how="right", op='contains')
geo_donald.head(5)
```

```python
geo_donald_mod.dropna(inplace=True)
useless_col = ["country", "city", "state", "state_code", "lat", "long"]
geo_donald_mod.drop(columns=useless_col, inplace=True)
# it's working yay

```

```python
print(geo_donald_mod[geo_donald_mod["SUB_REGION"] == "Pacific"])
```

```python
temp = geo_donald_mod
ax = usa.plot(color='white', edgecolor='black')
temp.plot(ax=ax, marker='o', markersize=12)
```
